### Numpy Array Operations

> Arithmatic Operations

```python
a1 = np.array([1,2,3,4,5])
a2 = np.array([1,2,3,4,5])
```

```python
a3 = a1 + a2
a3
```

```python
a4 = a1 ** a2
a4
```

> Broadcasting

```python
l = [1,2,3,4,5,6]
arr = np.array(l)
arr
```

```python
arr + 10
```

> Deep and Shallow Copy

```python
a = np.arange(1,21)
a
```

```python
slice = a[:5]
slice = slice * 10
slice
```

```python
a #deep copy
```

```python
b = a.copy()
b  #shallow cpopy
```

```python
b[0] = 1000
b
```
