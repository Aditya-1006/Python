# **Missing Values**

```python
import numpy as np
import pandas as pd
```

```python
data = {
    'A':[1,2,3,4,5],
    'B':[11,12,13,14,15],
    'C':[11,12,np.nan,np.nan,17],
    'D':[6,7,np.nan,8,9]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.isna()
```

```python
df.isna().count()
```

```python
df.isna().sum()
```

```python
df.isna().sum(axis = 1)
```

```python
df.isna().any()
```

```python
df.isna().any(axis=1)
```

- Removing Missing Values

```python
df
```

```python
df.dropna()    #removes all the rows with null values
```

```python
df.dropna(thresh=3)  #removes all the rows which have less than 3 non-NaN values
```

- Filling The Missing values

```python
df
```

```python
df.fillna(0)
```

*To frreze these values, Implace=True*

```python
values = {
    'A':00,
    'B':100,
    'C':200,
    'D':300
}
df.fillna(value=values)
```

```python
df.fillna(df.mean())
```
