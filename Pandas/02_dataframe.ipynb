```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Name" : ["Tejas", "Akash", "Suraj"],
    "Age" : [24, 23, 25],
    "City" : ["Nagpur", "Pune", "Akola"],
    "Salary" : [170000, 130000, 120000]
}

df = pd.DataFrame(data)
df
```
